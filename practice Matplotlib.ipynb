```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.show() #printing out the plot
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
```

```python
# functional method
plt.plot(x, y)
```

```python
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
############ multi-plot ###############
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

# Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
# need to put in one cell and run!
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig2 = plt.figure()
axes1 = fig2.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig2.add_axes([0.2, 0.5, 0.3, 0.3]) # insert set of axes!
axes1.plot(x, y)
axes1.set_title('Larger Plot')
axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3)
#axes.plot(x, y)
plt.tight_layout() # <------ fix the overlapping labels
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
# axes just a list of all the axes.
'''
# by iterating through the list
for current_ax in axes:
    current_ax.plot(x, y)
'''
# by indexing
axes[0].plot(x, y)
axes[0].set_title('First Plot')
axes[1].plot(y, x)
axes[1].set_title('Second Plot')

```

## Figure Size and DPI

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label = 'X-squared')
ax.plot(x, x**3, label = 'X-cubed')
# legends?
ax.legend(loc = 0)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize =( 5, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

## Part 3 of the practice

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color = '#FF8C00', linewidth = 5, alpha = 0.5) # default linewidth is 1.
# alpha argument is for transparency of the line! default also 1
# google search for RGB Hex Code
# for basic colors, they will be acceptable as strings
```

```python
######## line style
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color = 'purple', lw = 3, linestyle = '--') # abbr 'ls' for 'linestyle'
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color = 'purple', lw = 1, ls = ':', marker = 'o', markersize = '20', 
       markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color = 'purple', lw = 1, ls = '--')
# zoom in a little
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
