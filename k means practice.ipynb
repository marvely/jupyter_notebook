```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## generate artificial data

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples= 200, n_features= 2, centers = 4, 
                  cluster_std= 1.8, random_state = 101)
# created a tuple~
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c = data[1], cmap = 'rainbow')
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters= 4) 
```

```python
kmeans_model.fit(data[0])
```

```python
kmeans_model.cluster_centers_
```

```python
kmeans_model.labels_
# in real life, this step is the last step!
```

```python
# but here, we can compare the predictions~
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey= True, figsize = (10, 6))

ax1.set_title('K Means')
ax1.scatter(data[0][:, 0], data[0][:, 1], c = kmeans_model.labels_, cmap = 'rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:, 1], c = data[1], cmap = 'rainbow')
```

```python
kmeans_model_8 = KMeans(n_clusters= 8)
```

```python
kmeans_model_8.fit(data[0])
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c = kmeans_model_8.labels_, cmap = 'rainbow')
```

What KMeans thinks the data could be separated into 8 clusters~
